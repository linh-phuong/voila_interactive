```python
def plot_densitygrad(col_code, axes, alpha, beta):
    x, y = np.meshgrid(*axes)
    ct_dat = LV_compete_eq_coexist([x, y, alpha, beta])
    plt.contourf(x, y, ct_dat[0], cmap=col_code)
    plt.plot(axes[0], axes[0] * beta, c='k')
    direction = -1 if alpha < 0 else 1
    plt.plot(direction * axes[0], direction * axes[0]/alpha, c='k', linestyle='--')
    plt.xlim(-200, axes[0][-1])
    plt.ylim(-200, axes[1][-1])
```

```python
xrange = np.linspace(0,1000, 100)
yrange = np.linspace(0,1000, 100)
ipy.interact(plot_densitygrad, col_code=ipy.fixed('Blues'), axes=ipy.fixed((xrange, yrange)), 
             alpha=(-3.9476, 0.2, 0.5), beta=(0.8647, 2, 0.5))
```
